# Create 2D image for each molecule. Organize by inverision symmetry

```python
def makeImages(files, dataFolder, classFolder, myFrame):

    for i in range(len(files)):
        
        print(i, end=": ")

        molFile = str(files[i])
        m = Chem.MolFromMol2File(molFile)

        fileName = files[i].name
        fileName = fileName[:fileName.find('.')]
        
        row = myFrame.loc[myFrame['name'] == (' ' + fileName + ' ')]
        invSym = row['inversion_symmetry'].values[0]
        
        if (invSym == 0):
            invSymFolder = 'invSymFalse/'
            
        else:
            invSymFolder = 'invSymTrue/'
        
        absPath = dataFolder+classFolder+invSymFolder + fileName + '.png'

        print(absPath)

        try:
            smileInfo = Chem.MolToSmiles(m)
            m2 = Chem.MolFromSmiles(smileInfo)
            AllChem.Compute2DCoords(m2)
            Draw.MolToFile(m2, absPath)
        except:
            print("Failed to open/get smiles/draw... Deleting")
            os.remove(files[i])
            continue
            
        #Draw.MolToImage(m2)
```

```python
files = list(molPath16kTrain.iterdir())
dataFolder = '16kImages/'
classFolder = 'train/'

makeImages(files, dataFolder, classFolder, my16kFrame)
```

```python
files = list(molPath16kTest.iterdir())
dataFolder = '16kImages/'
classFolder = 'val/'

makeImages(files, dataFolder, classFolder, my16kFrame)
```

```python
files = list(molPath1kTrain.iterdir())
dataFolder = '1kImages/'
classFolder = 'train/'

makeImages(files, dataFolder, classFolder, my1kFrame)
```
